```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# File paths (replace with your actual paths)
file_path1 = 'API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv'
file_path2 = 'Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv'
file_path3 = 'Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv'

# Load the CSV files
data1 = pd.read_csv(file_path1, skiprows=4)  # Skipping metadata rows for the main population dataset
data2 = pd.read_csv(file_path2)
data3 = pd.read_csv(file_path3)

# Select relevant columns: 'Country Name' and '2020' population data
pop_2020 = data1[['Country Name', '2020']].dropna()

# Sort the data by population in descending order and select the top 20 countries for a clearer chart
pop_2020_sorted = pop_2020.sort_values(by='2020', ascending=False).head(20)

# Plotting the bar chart for population distribution in 2020
plt.figure(figsize=(12, 8))
plt.barh(pop_2020_sorted['Country Name'], pop_2020_sorted['2020'], color='skyblue')
plt.xlabel('Population in 2020')
plt.ylabel('Country')
plt.title('Top 20 Countries by Population in 2020')
plt.gca().invert_yaxis()  # To display the largest at the top
plt.tight_layout()

# Show the bar chart
plt.show()

# Now, let's plot a histogram of the population distribution for all countries in 2020
plt.figure(figsize=(10, 6))
plt.hist(pop_2020['2020'], bins=30, color='lightcoral', edgecolor='black')
plt.xlabel('Population in 2020')
plt.ylabel('Number of Countries')
plt.title('Histogram of Population Distribution Across Countries in 2020')
plt.tight_layout()

# Show the histogram
plt.show()
```
