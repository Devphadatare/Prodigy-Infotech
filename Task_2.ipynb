#Relationships between variables and identify patterns and trends in the data.

#1: Load Libraries and **Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File paths for the datasets
gender_submission_path = 'gender_submission.csv'
test_path = 'test.csv'
train_path = 'train.csv'

# Load the datasets
gender_submission_df = pd.read_csv(gender_submission_path)
test_df = pd.read_csv(test_path)
train_df = pd.read_csv(train_path)

```

#2: Clean the Data

```python
# Dropping unnecessary columns (Ticket, Cabin) and handling missing values
train_cleaned_df = train_df.drop(columns=['Ticket', 'Cabin'])
train_cleaned_df['Age'].fillna(train_cleaned_df['Age'].median(), inplace=True)
train_cleaned_df['Embarked'].fillna(train_cleaned_df['Embarked'].mode()[0], inplace=True)
```

#3: Visualize Missing Values Before and After Cleaning

```python
missing_values = train_df.isnull().sum()
missing_values_after_cleaning = train_cleaned_df.isnull().sum()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x=missing_values.index, y=missing_values.values, palette='coolwarm')
plt.title('Missing Values Before Cleaning')
plt.xticks(rotation=90)

plt.subplot(1, 2, 2)
sns.barplot(x=missing_values_after_cleaning.index, y=missing_values_after_cleaning.values, palette='coolwarm')
plt.title('Missing Values After Cleaning')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

#4: Data Exploration and Visualizations

#1. Survival Based on Gender

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=train_cleaned_df)
plt.title('Survival Based on Gender')
plt.show()
```

#2. Survival Based on Passenger Class

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Pclass', data=train_cleaned_df)
plt.title('Survival Based on Passenger Class')
plt.show()
```

#3. Age Distribution with Survival

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=train_cleaned_df, x='Age', hue='Survived', multiple="stack", kde=True)
plt.title('Age Distribution with Survival')
plt.show()
```

#4. Fare Distribution Across Classes (Bar Plot)

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Fare', data=train_cleaned_df, palette='coolwarm', estimator=sum)
plt.title('Fare Distribution Across Passenger Classes (Bar Plot)')
plt.show()
```

#5. Fare Distribution Across Classes (Histogram Subplots)

```python
plt.figure(figsize=(12, 8))
for i in sorted(train_cleaned_df['Pclass'].unique()):
    plt.subplot(3, 1, i)
    sns.histplot(train_cleaned_df[train_cleaned_df['Pclass'] == i]['Fare'], kde=False, bins=30)
    plt.title(f'Fare Distribution for Class {i}')
plt.tight_layout()
plt.show()
```
